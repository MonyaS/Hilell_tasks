```python
import numpy as np
```

Open file

```python
file_data = np.genfromtxt('iris.data', delimiter=',')
print(file_data)
```

Task â„–1

```python
file_data = np.delete(file_data, 4, 1)
print("Array after delete last column: \n", file_data)
```

Task â„–2

```python
#Optional to check operation
file_data = file_data.flatten()
print(file_data.shape)
```

```python
dem_col = file_data.ndim
if dem_col == 2:
    print("Number of array dimensions: ", dem_col)
    print("Array dimension: ", file_data.shape)
else:
    try:
        print("Number of array dimensions before: ", dem_col)
        print("Array dimension before: ", file_data.shape)
        file_data = file_data.reshape(150, 4)
        dem_col = file_data.ndim
        print("Number of array dimensions after: ", dem_col)
        print("Array dimension after: ", file_data.shape)
    except:
        print("This array is not suitable for transformation")
```

Task â„–3

```python
print("Mean of first column:", np.mean(file_data[0:, :1], axis=0))
print("Median of first column:", np.median(file_data[0:, :1], axis=0))
print("Standard deviation of first column:", np.std(file_data[0:, :1], axis=0))
```

Task â„–4

```python
index = np.arange(0, file_data.size)
np.random.shuffle(index)
file_data = file_data.flatten()
file_data[index[:20]] = np.nan
file_data = file_data.reshape(150, 4)
print("Array after replace 20 random items to 'nan': \n", file_data)
```

Task â„–5

```python
print("Indexes of 'nan' in first column: ", np.where(np.isnan(file_data[0:, :1]))[0])
```

Task â„–6

```python
mask = (file_data[0:, :1] < 5) & (file_data[0:, 3:] > 1.5)
temp = np.where(mask, file_data, 0).astype(bool)
print("Rows that match the conditions: \n", file_data[temp])
```

Task â„–7

```python
missing_bool = np.isnan(file_data)
file_data[missing_bool] = 0
print("Array after replace 'nan': \n", file_data)
```

Task â„–8

```python
unique_elem, counts = np.unique(file_data, return_counts=True)
print("Unique elements: ", unique_elem)
print("Counts: ", counts)
```

Task â„–9

```python
first, second = np.vsplit(file_data, 2)
print("First half: ", first, "\n\n\n\nSecond half:\n", second)
```

Task 10

```python
first_sort_indexes = first[:, 0].argsort()
first = first[first_sort_indexes]
second_sort_indexes = second[:, 0].argsort()[::-1]
second = second[second_sort_indexes]
print("First array after sorting:\n", first, "\nSecond array after sorting: \n", second)
```

Task 11

```python
file_data = np.vstack([first, second])
print("Array after concatenations: \n", file_data)
```

Task 12

```python
max_index = counts.argmax()
max_elem = unique_elem[max_index]
print("The most common element: ", max_elem, "\nIts quantity: ", counts[max_index])
```

Task 13

```python
def calculate(arr):
    mean = np.mean(arr)
    print(f"The mean of column: ",mean)
    return np.where(arr < mean, arr*2, (np.where(arr >= mean, arr/4,arr)))
```

Task 14

```python
file_data_copy=file_data.copy()
file_data_copy[:,3]=calculate(file_data[:,3]).flatten()
print("Numbers in 3rd columns in both arrays: ")
for was,now  in zip(np.nditer(file_data[:,3]),np.nditer(file_data_copy[:,3])):
        print("Was: ",was," Now: ",now)
print("Final array after processing of the 3rd column: ",file_data_copy)
```
